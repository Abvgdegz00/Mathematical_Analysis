**Задача 1**. (11 баллов)

В известном магазине техники представлено очень много экземпляров различной бытовой техники двух видов: с возможностью подключения к системе "Умный дом" и без такой возможности. Причем, известно, что эти два вида техники представлены в пропорции 1 : 3 соответственно. Случайным образом выбрали 7 экземпляров техники для розыгрыша среди участниц конкурса "Мисс Рукоделье - 2024".  Какова вероятность, что среди них не более двух экземпляров с возможностью подключения к системе "Умный дом"?

Подробно распишите решение и обоснуйте выбор распределения для моделирования сюжета задачи. Ответ округлите до трёх знаков.


По условию, мы имеем 7 независимых испытаний которые могут иметь два исхода: успех и неудача. Вероятность при каждом испытании не изменяется. Условия задачи подходят под биномиальное распределение.

Вероятность того, что выбранный экземпляр будет с возможностью подключения к системе "Умный дом", составляет:  

 $p = \frac{1}{1+3} = \frac{1}{4} = 0.25$

Вероятность того, что выбранный экземпляр не будет с возможностью подключения к системе "Умный дом", составляет:  

$q = 1 - p = 0.75$

Для вычисления используем кумулятивную функцию для биномиального распределения:  

$CDF(k)=P(X≤k)$

```python
# Количество испытаний
n = 7
# Вероятность, что в случайном испытании будет устройство с Умным домом
p = 0.25
# Количество успехов
k = 2
result = sp.stats.binom.cdf(k, n, p)
print("Вероятность составляет:", round(result, 3))
```

**Задача 2** (30 баллов)

Дан закон распределения количества публикаций в которой группе телеграм-каналов:

$$
\begin{array}{|c|c|c|c|c|c|c|}
\hline X \text { (число публикаций) } & 0 & 1 & 2 & 3 & 4 & 5 \\
\hline \mathrm{P} \text { (вероятность) } & ? & 0,3 & 0,3 & 0,2 & 0,1 & 0,1 \\
\hline
\end{array}
$$

Найдите пропущенную вероятность для X = 0, математическое ожидание и дисперсию. Решите задачу "вручную" и с помощью python.

__Пропущенная вероятность__

Чтобы найти вероятность при X = 0, используем тот факт, что сумма всех веротяностей составляет 1. В таком случае:  

$P(X = 0) = 1 - (0.3 + 0.3 + 0.2 + 0.1 + 0.1) = 0$

__Математическое ожидание__

Для расчета математического ожидания, требуется найти сумму произведений случайной величины на вероятность ее появления:  

$E(x) = 0*0 + 1*0.3 + 2*0.3 + 3*0.2 + 4*0.1 + 5*0.1 = 2.4$

__Дисперсия__

Для того, чтобы найти дисперсию, воспользуемся следующей формулой:  

$D(x) = E(x^2) - (E(x))^2$  

$E(x^2) = 0^2*0.0 + 1^2*0.3 + 2^2*0.3 + 3^2*0.2 + 4^2*0.1 + 5^2*0.1 = 0 + 0.3 + 1.2 + 1.8 + 1.6 + 2.5 = 7.4$  

$D(x) = 7.4 - 2.4^2 = 1.64$

```python
# Значения X и их вероятности
X_values = np.array([0, 1, 2, 3, 4, 5])
P_values = np.array([0, 0.3, 0.3, 0.2, 0.1, 0.1])

# Математическое ожидание
Mx = np.sum(X_values * P_values)

# Дисперсия
Dx = np.sum((X_values - Mx) ** 2 * P_values)

print("Математическое ожидание: ", Mx, "\nДисперсия: ", round(Dx,2))
```

Результат, полученный при ручном решении, совпадает с результатом, вычисленным программным методом.

**Задача 3** (25 баллов)


Компания приобрела 10 000 новых ноутбуков. Вероятность поломки одного ноутбука в течение месяца равна 0,0002. Предполагается, что поломки отдельных ноутбуков происходят независимо друг от друга.

**Вопрос 1**. Найдите вероятность того, что за месяц сломается ровно 1 ноутбук.
Решение объясните. (5 баллов)


Задачу можно представить как 10,000 независимых испытаний, в каждом из которых возможны два исхода: успех или неудача, с одинаковой вероятностью успеха в каждом испытании, равной 0,0002. Эти условия соответствуют биномиальному распределению.

```python
# Количество ноутбуков
n = 10000
# Вероятность поломки одного ноутбука
p = 0.0002
# Количество успехов
k = 1
# Вероятность того, что за месяц сломается ровно 1 ноутбук
P1 = sp.stats.binom.pmf(k, n, p)
print("Вероятность составляет:", round(P1,2))
```

**Вопрос 2**. Компания планирует ввести программу расширенной гарантии, по которой она будет бесплатно ремонтировать или заменять ноутбуки, сломавшиеся в течение месяца. Если общее количество поломок за месяц превысит 5 ноутбуков, компания понесет значительные дополнительные расходы.

а) Найдите вероятность того, что за месяц сломается более 5 ноутбуков.Решение объясните (10 баллов)

Для вычисления вероятности того, что сломается больше 5 ноутбуков за месяц, можно использовать SF (Survival Function):  

$SF(k)=P(X>k)$

```python
P2 = sp.stats.binom.sf(5, n, p)
print("Вероятность составляет:", round(P2, 3))
```

б) Если стоимость ремонта или замены одного ноутбука составляет 500 долларов, вычислите математическое ожидание общих расходов компании на поломки за один месяц. (10 баллов)

```python
# Стоимость ремонта одного ноутбука
cost = 500
# Математическое ожидание количества поломок за месяц
Mx_broke = n * p
# Математическое ожидание общих расходов
Mx_total = cost * Mx_broke
print("Математическое ожидание общих расходов:", Mx_total)
```
